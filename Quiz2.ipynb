### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year', color='r', marker='o', markerfacecolor='yellow', linestyle='--', linewidth=4)
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('company profit')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo culumn (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['toothpaste']
faceWashSalesData   = df ['facewash']

plt.bar([a for a in monthList], faceCremSalesData, width= 0.25, label = 'Ftoothpaste sales data', align='edge')
plt.bar([a for a in monthList], faceWashSalesData, width= -0.25, label = 'shampoo sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Ftoothpaste and shampoo sales data')
plt.show()
```

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
# create plot
sns.countplot(x = 'survived', hue = 'sex', data = titanic, palette = 'magma')
plt.title('Survivors')
plt.show()
```

### 10+ - what function can we use instead of kind="line" in seaborn?
