```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df=pd.read_csv('supermarket.csv')
df.columns = df.columns.str.replace(' ', '')
df.head()
```

```python
#one
df['Product'].nunique()
```

```python
#Two
df2=df.sort_values(by='Date')
```

```python
df2['count_pro'] = df2['Product'].str.count(',').add(1).fillna(0) 
df2.head()
```

```python
df2=np.round(df2.groupby(df2['Date'])['count_pro'].sum().mean())
df2
```

```python
#Three
total=df['Product'].value_counts()
df3=total.to_frame()
df3.tail(4)
```

```python
#Four
df['Date'] = pd.to_datetime(df['Date'])
df['day_of_week'] = df['Date'].dt.day_name()
df4=df.sort_values(by='CustomerId')
mask=(df['Date']=='2020')
df4=df4.loc[mask]
df4.to_csv('new.csv')
df4=pd.read_csv('new.csv',sep=',')
total=df4['CustomerId'].value_counts()
df4=total.to_frame()
df4.head()
```

```python
#Five
total=df.groupby('Date')['day_of_week'].value_counts().sort_values(ascending=False)
df5=total.to_frame()
df5.head(3)
```
