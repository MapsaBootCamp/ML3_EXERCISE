```python
import pandas as pd
import numpy as np
```

```python
cpm_p_table = pd.read_csv('quize_pandas.zip', compression='zip')
```

Questions and Results

با استفاده از دادگان ارائه شده محاسبه کنید هر مشتری چند درصد از هر محصول را خریده است. (با دقت دهگان درصد)

```python
cpm_p_table['Quantity'] = cpm_p_table.loc[:,['Volume', 'Count']].max(axis=1)
```

```python
cpm_p_table['Total'] = 0
mask = 0
for index, row in cpm_p_table.iterrows():
    mask = cpm_p_table.Product == row['Product']
    cpm_p_table['Total'][index] = cpm_p_table[mask]['Quantity'].sum()
    
cpm_p_table['Percent'] = cpm_p_table['Quantity']*100/cpm_p_table['Total']

df = cpm_p_table.groupby(['Customer', 'Product'])
df = np.around(df.sum().reset_index(), 2)
df
```

```python

df.pivot(index = 'Customer', columns=['Product'],values='Percent').fillna(0).astype(str).add('%')
```
