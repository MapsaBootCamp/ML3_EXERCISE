```python
from hazm import stopwords_list
stopwords_list()
```

```python
import wordcloud
from wordcloud import WordCloud
from wordcloud_fa import WordCloudFa
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
```

## rohani

```python
file = open("rohani.txt") 
rohani_speech =file.read()

wordcloud = WordCloudFa(font_path="Mj_Vanilla_0.ttf",width = 800, height = 800, background_color ='Black',stopwords=stopwords_list(), min_font_size = 10).generate(str(rohani_speech))
 
plt.figure(figsize = (5, 5))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
