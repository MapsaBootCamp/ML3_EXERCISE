```python
data['Count'] = data['Count'].dropna(0)
data['percent']=data['Count']/data['Count'].sum()*100
data
```

```python
data1 = data['count_total'] = data.groupby(['Product','Model'])['Count'].transform('sum')
data1 = data['volume_total'] = data.groupby(['Product','Model'])['Volume'].transform('sum')
data1 = pd.pivot_table(data, values=('Count','Volume','count_total','volume_total'), index=['Customer', 'Product','Model'])
data1
```
