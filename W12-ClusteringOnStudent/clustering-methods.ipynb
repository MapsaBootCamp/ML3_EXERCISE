```python
students_df = pd.read_excel("data_students.xls", sheet_name="Training_Data")
students_df.head()
```

```python
students_df = students_df.loc[:, ["STG", "SCG", "STR", "LPR", "PEG"]]
students_df.describe()
```

```python
fig = plt.figure(figsize=(9,6))
X = students_df[['STG' , 'PEG']].values
plt.scatter(X[:,0], X[:,1])
plt.xlabel('STG') , plt.ylabel('PEG')
```

### KMeans Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
```

```python
inertia = []
for cl_num in range(1 , 10):
    km = KMeans(n_clusters = cl_num, max_iter=100, n_init='auto', random_state= 100)
    km.fit(X)
    inertia.append(km.inertia_)
```

```python
fig = plt.figure(figsize=(5, 2))

plt.plot(np.arange(1 , 10) , inertia , 'o')
plt.plot(np.arange(1 , 10) , inertia , '-')
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
```

```python
k = 3
km_cluster = KMeans(n_clusters = k , n_init = 'auto',max_iter=200, random_state= 120)
km_cluster.fit(X)
labels = km_cluster.labels_
centroids = km_cluster.cluster_centers_
```

```python
fig = plt.figure(figsize=(9, 6))

plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centroids[:,0] ,centroids[:,1], color='red', marker="*")
plt.title("Find Unknown groups in students Data", fontsize=10)
```

```python
davies_bouldin_score(X, labels)
```

### DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.05,min_samples=5,metric='euclidean')
dbscan.fit(X)
labels = dbscan.labels_
```

```python
fig = plt.figure(figsize=(9, 6))

plt.scatter(X[:,0], X[:,1], c=labels)
```

```python
davies_bouldin_score(X, labels)
```

### SpectralClustering

```python
from sklearn.cluster import SpectralClustering
```

```python
sc=SpectralClustering(n_clusters=4,assign_labels='discretize', random_state=0)
sc.fit(X)
labels = sc.labels_
```

```python
fig = plt.figure(figsize=(9, 6))

plt.scatter(X[:,0], X[:,1], c=labels)
```

```python
davies_bouldin_score(X, labels)
```

### GaussianMixture

```python
from sklearn.mixture import GaussianMixture
```

```python
ini = ["kmeans", "random_from_data", "k-means++", "random"]
gmm = GaussianMixture(n_components=4, init_params=ini[2], tol=1e-9, max_iter=2000, random_state=0)
gmm.fit(X)
means = gmm.means_
```

```python
means
```

```python
colors = ["navy", "turquoise", "cornflowerblue", "darkorange"]
for i, color in enumerate(colors):
        data = X[gmm.predict(X) == i]
        plt.scatter(data[:, 0], data[:, 1], color=color, marker="x")
```

### Birch

```python
from sklearn.cluster import Birch
```

```python
b=Birch(threshold=0.01, n_clusters=4)
b.fit(X)
labels = b.labels_
```

```python
fig = plt.figure(figsize=(9, 6))

plt.scatter(X[:,0], X[:,1], c=labels)
```
