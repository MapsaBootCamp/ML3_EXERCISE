# 1- SVM

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
```

```python
X = df.drop('Class', axis = 1).values
y = df['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
SVM_Model = svm.SVC(probability=True)
SVM_Model.fit(X_train, y_train)
y_pred = SVM_Model.predict(X_test)
predicted_train = SVM_Model.predict(X_train) 
training_accouracy = accuracy_score(y_train, predicted_train) 
Accuracy_score=accuracy_score(y_test, y_pred)
Precision_Score=precision_score(y_test, y_pred, average='macro')
print(f"Training phase accuracy: {training_accouracy}")
print(f"Testing phase accuracy: {Accuracy_score}")
print(f"precision score: {Precision_Score}")
print('Confusion matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
y_pred_proba = SVM_Model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 2- Decision Tree

```python
from sklearn import tree
```

```python
X = df.drop('Class', axis = 1).values
y = df['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
tree_Model = tree.DecisionTreeClassifier(max_depth=5, min_samples_leaf=7)
tree_Model.fit(X_train, y_train)
y_pred = tree_Model.predict(X_test)
predicted_train = tree_Model.predict(X_train) 
training_accouracy = accuracy_score(y_train, predicted_train) 
Accuracy_score=accuracy_score(y_test, y_pred)
Precision_Score=precision_score(y_test, y_pred, average='macro')
print(f"Training phase accuracy: {training_accouracy}")
print(f"Testing phase accuracy: {Accuracy_score}")
print(f"precision score: {Precision_Score}")
print('Confusion matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
y_pred_proba = tree_Model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 4- Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
```

```python
X = df.drop('Class', axis = 1).values
y = df['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
#GNB_Model = GaussianNB()
param_grid_nb = {'var_smoothing': np.logspace(0,-9, num=100)}
GNB_Model = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid_nb, verbose=1, cv=10, n_jobs=-1)
GNB_Model.fit(X_train, y_train)
y_pred = GNB_Model.predict(X_test)
predicted_train = GNB_Model.predict(X_train) 
training_accouracy = accuracy_score(y_train, predicted_train) 
Accuracy_score=accuracy_score(y_test, y_pred)
Precision_Score=precision_score(y_test, y_pred, average='macro')
print(f"Training phase accuracy: {training_accouracy}")
print(f"Testing phase accuracy: {Accuracy_score}")
print(f"precision score: {Precision_Score}")
print('Confusion matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
print(GNB_Model.best_estimator_)
```

```python
y_pred_proba = GNB_Model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
