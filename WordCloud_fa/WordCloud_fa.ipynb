```python
from wordcloud import WordCloud
from hazm import stopwords_list
from wordcloud_fa import WordCloudFa
from PIL import Image
```

# Ahmadinejad

```python
filee = open("ahmadinejad",'r') 
ahmadinejad = str(filee.read())
print(type(ahmadinejad))
```

```python
word = WordCloudFa(font_path="B-NAZANIN.TTF",width = 500, height = 300,stopwords=stopwords_list())
word_cloud = word.generate(ahmadinejad)
image = word_cloud.to_image()
image
```
