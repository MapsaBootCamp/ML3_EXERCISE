```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
nltk.download()
nltk.download('stopwords')
from nltk.corpus import stopwords
from wordcloud import WordCloud
from hazm import stopwords_list

```

```python
Ahmadi=open ("ahmadinejad",'r')
Katamy=open ("khatami",'r')
Rohany=open ("rohani",'r')
Raesi=open ("raesi",'r')
Ahmadi= Ahmadi.read()
Katamy= Katamy.read()
Rohany=Rohany.read()
Raesi=Raesi.read()
presidents=[Ahmadi,Ahmadi,Rohany,Raesi]

```

```python
for president in presidents:
    
    WordCloud1=WordCloud(font_path='B Zar_0.ttf',width = 800, height = 800,background_color ='white',\
     stopwords=stopwords_list(),min_font_size = 10).generate(president)
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(WordCloud1)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    plt.show()
```
