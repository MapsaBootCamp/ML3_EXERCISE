```python
import numpy as np
import pandas as pd

movies = pd.read_csv("imdb_movie_rank.csv")
movies.head()
```

**حذف داده های null**

```python
movie = movies.dropna(subset = ["content_rating"])
movie
```

**مرتب کردن بر اساس زمان و ذخیره 50 تای اول**

```python
sorted = movies.sort_values("duration", ascending= False)
sort = sorted.head(50)
sort.to_csv('Long_time.csv')
sort
```

**تابعی که اسم بازیگر رو میگیرد و فیلم های آن را برمیگردونه**

```python
def actors(actor):
  info = movies.loc[movies['actors_list'].str.contains(actor)]  
  return info
print(actors('Robert De Niro'))
```

**میانگین امتیاز فیلم بر اساس ژانر**

```python
movies.groupby('genre').star_rating.mean()
```

**اضافه کردن ستون 0و1**

```python
movies["new"]=movies["duration"].apply(lambda n: 0 if n >= 150  else 1)
movies.head()
```

**گروه بندی بر اساس ژانر و جداکردن 3تای برتر آن**
