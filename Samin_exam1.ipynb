## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Company_sales = pd.read_csv('company_sales_data.csv')

Company_sales.plot('month_number','total_profit')
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
#Company_sales.plot('month_number','total_profit',ylabel = 'profit', xlabel = 'Month number', title = 'company profit', xticks=range(1,13), yticks =(100000, 200000, 300000, 400000, 500000))
plt.plot(Company_sales['month_number'],Company_sales['total_profit'])
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('Company profit')
plt.xticks(range(1,13))
plt.yticks((100000, 200000, 300000, 400000, 500000))
plt.show()
```

### style change

#### 5 - make your plot appearance like below(5 parameters should change)

```python
plt.plot(Company_sales['month_number'],Company_sales['total_profit'],marker= "o",markerfacecolor='yellow',linestyle='dashed', color = 'red',linewidth=5.0)
plt.xlabel('Month number')
plt.ylabel('profit')
plt.title('Company profit')
plt.xticks(range(1,13))
plt.yticks((100000, 200000, 300000, 400000, 500000))
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
Company_sales.plot(x="month_number", y=["toothpaste","shampoo" ], kind="bar")
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Ftoothpaste and shampo sales data")
plt.grid(True, axis='y',linestyle = '--')
plt.grid(True, axis='x', linestyle = '--')
plt.show()
```

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
titanic = sns.load_dataset('titanic')
sns.catplot(x ="survived", hue ="sex", kind ="count", data = titanic,palette=sns.color_palette("tab10")).set(title='Survivors')
```
