```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from __future__ import unicode_literals
from hazm import *
import wordcloud
from hazm import stopwords_list
from wordcloud import WordCloud
from wordcloud_fa import WordCloudFa
from nltk.tokenize import word_tokenize
```

```python
lAddress =open('khatami', 'r',encoding="utf8")
str=lAddress.read()
lAddress =open('raesi', 'r',encoding="utf8")
strraesi=lAddress.read()
lAddress =open('rohani', 'r',encoding="utf8")
strrohani=lAddress.read()
```

```python
word_tokens = word_tokenize(strrohani)
wc = WordCloudFa()
```

```python
stop=stopwords_list()
stop.append('مى')
stop.append('،')
stop.append('.')
stop.append('.')
stop.append('های')
stop.append('ها')
stop.append(')')
stop.append('«')
stop.append('»')
stop.append('است')

```

```python
filtered_text = [w for w in word_tokens if not w.lower() in stop]
df=pd.DataFrame(filtered_text)

dfm=df.groupby(by=0).size().nlargest(30)
dfm.head(3)
```

```python
strrohani=" ".join(filtered_text)
```

```python
wordcloud = WordCloudFa( width = 800, height = 800, background_color ='white',stopwords=stop, min_font_size = 10).generate(str)
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
print('−−−−−−−−−−−−−−khatami−−−−−−−−−−−−−−−−') 
plt.show()
```
