#### 1- read https://quera.org/problemset/125361/ train dataset comments and show it with WordCloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from hazm import stopwords_list
```

```python
train_df = pd.read_csv('train.csv')
comment_column = " ".join(train_df['comment'])
wordcloud = WordCloud(font_path='B-NAZANIN.TTF',stopwords=stopwords_list(),collocations=False,width = 800, height = 800,background_color ='white').generate(comment_column)
# plot the WordCloud image                      
plt.figure(figsize = (4, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
