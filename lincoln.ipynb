```python
with open('lincoln-last-speech.txt',"r") as f:
    lines = f.readlines()
```

```python
from nltk.corpus import stopwords
```

```python
lines1=str(lines)
```

```python
a=stopwords.words("english")
```

```python
from nltk.tokenize import word_tokenize

word_tokens = word_tokenize(lines1[1:-1])

filtered = [w for w in word_tokens if not w.lower() in a]
word_tokens
```

```python
filtered
```

```python
ss=[]
for i in filtered:
    if i!= "," and i !=";" and i!="." and i!="'":
        ss.append(i)
```
