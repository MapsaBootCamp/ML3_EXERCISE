```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

supermarket = pd.read_csv('supermarket.csv')
supermarket
```

*تعداد محصولات یکتا در کل دادگان، چندتاست؟*

```python
print(len(supermarket["Product"].unique()))
```

```python
pd.pivot_table(supermarket, values=('Date'), index=['Customer Id', 'Product'], aggfunc='first')
```

*چهار محصولی که کمتر از بقیه در سبد مشتریان قرار گرفته‌اند، کدامند؟*

```python
supermarket['Product'].value_counts()
```
