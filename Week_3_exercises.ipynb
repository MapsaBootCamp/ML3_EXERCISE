#  اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند









```python
index_numbers=np.where(numbers<8.5)
new1=numbers[index_numbers]
print(np.shape(new1))
def f(x):
  y=(10+(x/10))
  return y 
g1 = np.vectorize(f)
result=g1(new1)
print(result)
print('Max:',result.max())
```

# اگر عددی بیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند 

```python
index_numbers=np.where(numbers>=8.5)
new2=numbers[index_numbers]
print(np.shape(new2))
def f(x):
  y=-0.0051*(x**2)+0.8586*x+3.4332
  return y
g1 = np.vectorize(f)
result2=g1(new2)
print(result2)
print('Max:',result2.max())

```

# توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید

```python
len_down=len(np.array(result2))
top_new=round(len_down*0.25)

top_new_twenty_five = np.random.choice(result2, top_new, replace=False)
top_new_twenty_five
```
