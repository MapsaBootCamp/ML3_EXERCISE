```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import hazm
from hazm import stopwords_list
from wordcloud_fa import WordCloudFa
```

```python
with open('ahmadinejad', 'r',encoding="utf8") as file:
    ahmadinejad = file.read()
with open('khatami', 'r',encoding="utf8") as file:
    khatami = file.read()
with open('raesi', 'r',encoding="utf8") as file:
    raesi = file.read()
with open('rohani', 'r',encoding="utf8") as file:
    rohani = file.read()
```

```python
wc = WordCloudFa(background_color="white",persian_normalize=True)
```

```python
wordcloud = wc.generate(ahmadinejad)
# plot the WordCloud image                      
plt.figure(figsize = (4, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
