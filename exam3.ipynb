### 6- make the plaot with bokeh like titanic.html file x axis is age y axis is survived and bolding of circles are according to Fare when you hover each circle it should show some extra informations

```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row 
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()
```

```python
from bokeh.transform import factor_cmap
from bokeh.models import LinearInterpolator
```

```python
titanic_df=pd.read_csv('titanic.csv')
titanic_df.head(2)
```

```python
index_cmap = factor_cmap('Sex', palette=['red', 'blue'], 
                         factors=sorted(titanic_df.Sex.unique()))
```

```python
size_mapper = LinearInterpolator(
    x=[1,50],
    y=[1,50]
)
```

```python
p = figure(plot_width=600, plot_height=450, title = "Survived people",toolbar_location=None,
          tools="hover")
p.add_tools(HoverTool(show_arrow=False, line_policy='next', tooltips=[('سن',' @Age'),('جنس','@Sex'),('Fare','@Fare'),('اسم','@Name'),('کلاس','@Pclass')]))
    
    

p.scatter('Age','Survived',source=titanic_df,fill_alpha=0.6, fill_color='blue',size={'field':'Fare','transform': size_mapper})
p.legend.location = "center_left"
show(p)
```
