## mathplotlib

#### 1 - read company_sales_data.csv with pandas

#### 2 - make a plot with total_profit as x axis and month_number columns as y axis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('company_sales_data.csv')
df.head(2)
```

#### 3 - add "profit" as x label and "Month number" as y label  "company profit" as title  


#### 4 - each month_number should show in x axis and y axis should be [100000, 200000, 300000, 400000, 500000]

```python
plt.rcParams.update({'font.size': 6})
plt.figure(figsize=(5,3))

plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('month')
plt.ylabel('profit')
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit')
plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.plot(df['month_number'], df['total_profit'], color="red" , linestyle='dashed', marker="o", markersize=6 ,markerfacecolor='y',linewidth=2) # 
plt.xlabel('month')
plt.ylabel('profit')
plt.xticks(df['month_number'])
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit')
plt.show()
```

### bar plot 
#### 6 - create plot with toothpaste and shampoo column (if you can make background girdy)
#### use width and align='edge' to print bars beside each others not top of each others)

```python
plt.figure(figsize=(5,3))

plt.bar(df['month_number']+ 0.25, df['toothpaste'] , color ='royalblue', width = 0.25,  label ='toothpaste')
plt.bar(df['month_number'] , df['shampoo'] , color ='lightcoral', width = 0.25,   label ='shampoo')
plt.xlabel('month')
plt.ylabel('profit')
plt.xticks(df['month_number'])
# plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit')
plt.legend(['Sales toothpaste', 'Sales shampoo'])
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.6)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv('train.csv')
df.head(2)
```

#### 9 - load titanic dataset with seaborn  and plot the count of survivors according to sex

```python
df3 = df.loc[:,['Sex','Survived']]
df3.head()
df2=df3.groupby(['Survived','Sex']).size().reset_index()
df2.rename(columns={0:'count'}, inplace=True)
plt.figure(figsize=(5,3))
sns.barplot(x="Survived",y ="count", hue="Sex"  , data=df2)
```
