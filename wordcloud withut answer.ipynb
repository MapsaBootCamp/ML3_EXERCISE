#### 1- read https://quera.org/problemset/125361/ train dataset comments and show it with WordCloud

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
import hazm
```

```python
comments = " ".join(df["comment"])
```

```python
word_tokens = hazm.word_tokenize(comments)
```

```python
stop_words = hazm.stopwords_list()
```

```python
cleaned_words = [item for item in word_tokens if item not in stop_words]
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white',font_path = "B-NAZANIN.TTF",
                      min_font_size = 10).generate(" ".join(cleaned_words))

plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
