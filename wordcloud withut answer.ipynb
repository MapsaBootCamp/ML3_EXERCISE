#### 1- read https://quera.org/problemset/125361/ train dataset comments and show it with WordCloud

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from __future__ import unicode_literals
from hazm import *
import wordcloud
from hazm import stopwords_list
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize
```

```python
df= pd.read_csv('train.csv')  
strc=' '.join(df['comment'].tolist())
```

```python
wordcloud = WordCloud( width = 800, height = 800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(strc)
 # font_path="/home/foad/B-NAZANIN.TTF",
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
stop.append('')
stop.append('')
stop.append('.')
stop.append('.')
stop.append('')
stop.append('')
stop.append(')')
stop.append('')
stop.append('')
stop.append('')
stop.append('!')
```

```python
word_tokens = word_tokenize(strc)
filtered_text = [w for w in word_tokens if not w.lower() in stop]
strc=" ".join(filtered_text)
```
