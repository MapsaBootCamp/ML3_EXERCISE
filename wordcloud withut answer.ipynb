#### 1- read https://quera.org/problemset/125361/ train dataset comments and show it with WordCloud

```python
data = pd.read_csv('train.csv')
data.head(3)

lAddress=' '.join(data['comment'])
```

```python
wordcloud = WordCloud(font_path="/home/cmos/maryam/B-NAZANIN.TTF" ,width = 800, height = 800, 
                      background_color ='black', 
                      colormap='Set2',
                      stopwords=stopwords_list(),
                      min_font_size = 10).generate(lAddress)
 
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
