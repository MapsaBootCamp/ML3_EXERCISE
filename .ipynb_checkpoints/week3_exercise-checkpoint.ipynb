		<p></p>
        برای این مجموعه داده ها قصد داریم با دو فرمول زیر مقادیر آنها را تغییر دهیم:
<br>
۱. اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند:
<br>
$$ y = 10 + \frac{x}{10} $$
<br>
۲. اگر عددیبیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند :
<br>
$$ y =  -0.0051x^{2} + 0.8586x + 3.4332 $$
<br>
توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید
</div>

```python
new_numbers=np.array(list(map(lambda x:10+(x/10) if x<8.5 else (-0.0051*(x**2))+(0.8586*x)+3.4332,numbers)))
new_numbers=new_numbers[(new_numbers<=23)]

        
```

```python
new_numbers=np.sort(new_numbers)
top_new_twenty_five = new_numbers[(3*len(new_numbers)//4)+1:]
```

 25 درصد داده های کوچکتر آرایه اول را در آرایه زیر بریزید
</div>

```python
down_old_twenty_five = new_numbers[:(len(new_numbers)//4)]
ones-np.ones()
```

<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
      <p></p>
        داده های متناظر دو آرایه به دست آمده را از یکدیگر کسر کنید و در یک آرایه سه بعدی که اندازه ابعاد سوم و دوم آن یک میباشد بریزید
</div>

```python
subtract=top_new_twenty_five-down_old_twenty_five

def myfunc(subtract):
    ones=np.ones((1,len(subtract),3))
    for i in range (0,len(subtract)):
        ones[0,i,0]=subtract[i]
    return ones
print(myfunc(subtract))

```

```python
name=input('give me the name\n')

def find_movies(name):
    name=name.lower()
    actors=movies['actors_list']
    actors=actors.str.lower().str.find(name)
    actor=actors.where(actors!=-1).dropna()
    return count_movies(actor)
def count_movies(actor):
    counter=0
    indexes=actor.index
    for index in indexes:
        if movies.at[index,'star_rating']>=7:
            counter+=1
    return counter
        
print(find_movies(name))

```
