```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords  
import wordcloud
```

```python
nltk.download('stopwords')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML3/Exercises/quizes/quize.3/train.csv')
```

```python
comment_words=" ".join(df['comment'])

Word_cloud=WordCloud(font_path="/content/drive/MyDrive/Colab Notebooks/ML3/Exercises/quizes/quize.3/B-NAZANIN.TTF", width=800 , height=800,
                     background_color='white',stopwords=stopwords.words('english'),
                     min_font_size=10).generate(comment_words)
                    
plt.figure(figsize = (3,3) , facecolor = None)
plt.imshow(Word_cloud)
plt.axis('off')
plt.tight_layout(pad=0)

plt.show()
```
