```python
import numpy as np
import pandas as pd
from numpy.linalg import svd
```

```python
df1= pd.read_csv('10000.txt',sep='\t',names=['user_id','song_id','repeated'])
df2= pd.read_csv('song_data.csv')
```

```python
# there is no NaN
df1.isna().values.any()
```

```python
df=df2.merge(df1,how='inner', on='song_id')
```

```python
df
```

```python
item_col = 'song_id'
user_col = 'user_id'
freq_col = 'repeated'

mat = df.pivot_table(
    index = user_col, 
    columns = item_col, 
    values = freq_col, 
    fill_value=0,
    aggfunc='sum'
    
)
```

ابده اصلی برای مقادیری که نداریم این است که :
از روش semi-supervised استفاده کنیم
بدین صورت که تمامی افراد را برای یک اهنگ تاوگت قرار دهیم و از بقیه ستون ها به عنوان فیچر استفاده کنیم و داده های جدید را پیش بینی کنیم و این کار را برای تمامی اهنگ ها انجام دهیم 
اما به دلیل زیاد بودن اهتگ ها این کار از لحاظ فنی قابل انجام نمی باشد و در نتیجه همینجوری ادامه می دهیم
