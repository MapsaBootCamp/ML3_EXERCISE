```python
import pandas as pd
from wordcloud_fa import WordCloudFa
import matplotlib.pyplot as plt
from hazm import stopwords_list
from string import punctuation
```

# روحانی

```python
speech = " ".join(pd.read_csv("rohani.txt",header=None)[0])
speech = "".join([char for char in speech if (char.isalpha()) or (char == " ") or (char.isdigit())])
```

```python
word_cloud = WordCloudFa(width = 1800, height = 1800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(speech)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(word_cloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
