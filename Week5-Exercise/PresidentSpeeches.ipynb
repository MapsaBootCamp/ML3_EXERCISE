```python
import nltk
import matplotlib.pyplot as plt
import pandas as pd
import wordcloud
from wordcloud import WordCloud
```

```python
with open('khatami.txt','r',encoding="utf8") as f:
        khatami_text=f.read()
with open('ahmadinejad.txt','r',encoding="utf8") as f:
        ahmadinejad_text=f.read()
with open('rohani.txt','r',encoding="utf8") as f:
        rohani_text=f.read()
with open('raesi.txt','r',encoding="utf8") as f:
        raesi_text=f.read()
```

```python
from hazm import stopwords_list
st=stopwords_list()+['می','های','ها']
```

```python
wordcloud = WordCloud(font_path= "B-NAZANIN.TTF", width = 800, height = 800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(khatami_text)
 
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.title('خاتمی')
plt.tight_layout(pad = 0)
 
plt.show()

wordcloud = WordCloud(font_path= "B-NAZANIN.TTF", width = 800, height = 800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(ahmadinejad_text)
 
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.title('احمدی نژاد')
plt.tight_layout(pad = 0)
 
plt.show()

wordcloud = WordCloud(font_path= "B-NAZANIN.TTF", width = 800, height = 800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(rohani_text)
 
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.title('روحانی')
plt.tight_layout(pad = 0)
 
plt.show()

wordcloud = WordCloud(font_path= "B-NAZANIN.TTF", width = 800, height = 800, 
                      background_color ='white',stopwords=stopwords_list(),
                      min_font_size = 10).generate(raesi_text)
 
plt.figure(figsize = (3, 3), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.title('رئیسی')
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
from wordcloud_fa import WordCloudFa

wodcloud = WordCloudFa(no_reshape=False, persian_normalize=True, include_numbers=False, collocations=False, width=800, height=400)
wc = wodcloud.generate(khatami_text)

plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.title('خاتمی')
plt.tight_layout(pad = 0)
 
plt.show()

wodcloud = WordCloudFa(no_reshape=False, persian_normalize=True, include_numbers=False, collocations=False, width=800, height=400)
wc = wodcloud.generate(ahmadinejad_text)

plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.title('احمدی نژاد')
plt.tight_layout(pad = 0)
 
plt.show()

wodcloud = WordCloudFa(no_reshape=False, persian_normalize=True, include_numbers=False, collocations=False, width=800, height=400)
wc = wodcloud.generate(rohani_text)

plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.title('روحانی')
plt.tight_layout(pad = 0)
 
plt.show()

wodcloud = WordCloudFa(no_reshape=False, persian_normalize=True, include_numbers=False, collocations=False, width=800, height=400)
wc = wodcloud.generate(raesi_text)

plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.title('رئیسی')
plt.tight_layout(pad = 0)
 
plt.show()
```
