```python
import numpy as np
import pandas as pd
from shapely.geometry import Point, LineString, Polygon
import matplotlib.pyplot as plt
from shapely import geometry
import numpy as np
import pandas as pd
import geopandas as gpd
%matplotlib inline

state_df = pd.read_csv('usa_state.txt')
state_df
```

```python
import numpy as np
import pandas as pd

arrests_df = pd.read_csv('usa_arrests.txt')
arrests_df.head()
```

```python
merged_df = pd.merge(state_df, arrests_df, left_on='region', right_on='state', how='left')
ordered_df = merged_df.sort_values('order').reset_index(drop=True)
ordered_df.head()
```

```python
sdata_df = ordered_df[['long' , 'lat' , 'group' , 'region' , 'Murder' , 'Assault' , 'UrbanPop' , 'Rape']]
```

```python
molten_data = pd.melt(sdata_df, id_vars=['long' , 'lat' , 'group' , 'region'])
molten_data
```

```python
molten_data['point'] = list(zip(molten_data.long, molten_data.lat))
molten_data['geometry'] = molten_data['point'].apply(Point)
molten_data.head()
```

```python
import geopandas as gpd
cities = gpd.GeoDataFrame(molten_data, geometry='geometry')
cities.head()
```

```python
[fig,ax] = plt.subplots(1, figsize=(12,8))
cities.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.5, legend=True, ax=ax, scheme='equal_interval')
ax.axis('off')
```
