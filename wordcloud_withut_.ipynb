#### 1- read https://quera.org/problemset/125361/ train dataset comments and show it with WordCloud

```python
import pandas as pd
from wordcloud import WordCloud
from hazm import stopwords_list
from wordcloud_fa import WordCloudFa
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
filee = pd.read_csv('train.csv')
f=' '.join(filee['comment'])
```

```python
word = WordCloud(font_path="B-NAZANIN.TTF",width = 500, height = 300,stopwords=stopwords_list())
word_cloud = word.generate(f)
image = word_cloud.to_image()
image
```

```python
titanic_df=pd.read_csv('titanic.csv')
titanic_df
```

```python
df_mean=np.round(titanic_df.groupby(['Pclass']).agg({'Survived':'mean'})*100)
      
feature_data = df_mean.index
feature_count=df_mean['Survived']

plt.figure(figsize = (9,3))
plt.bar(feature_data,feature_count)
plt.xticks(feature_count.index, feature_count.index.values)
plt.title('Percentage of Survivor peaple according to class')
plt.show()
```
