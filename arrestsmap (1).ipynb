```python
# import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

import PIL.ImageDraw as ImageDraw
import PIL.Image as Image

# from shapely.geometry import Point

%matplotlib inline

df=pd.read_csv("usa_state.csv")
arrest=pd.read_csv("usa_arrests.csv", index_col=5)
arrest
df=df[df.region != 'district of columbia']
```

```python
image = Image.new("RGB", (640, 480), color=(222,222,244))
draw = ImageDraw.Draw(image)

a=7
df['coor'] = df.apply(lambda x: (((150+x.long)*a ,(-x.lat+70)*a)), axis=1)
states=set(df['region'])

for i in states:
    # print(i,c)
    murder=arrest.loc[i,'Murder']
    zarib_rang=int((murder+2)*10)
    df2=df[df['region']==i]
    points = tuple(list(df2['coor']))
    draw.polygon(points, fill=(5*zarib_rang,11*zarib_rang,8*zarib_rang), outline=20, width=1)
image.show()
df2


```
