```python
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from hazm import stopwords_list
stopwords_list
from PIL import Image
```

```python

with open ("ahmadinejad") as file:
    ahmadinejad = file.read()
mask = np.array(Image.open("/home/cmos/Downloads/week5/mahmadi.jpeg"))    
wordcloud = WordCloud(font_path="/home/cmos/maryam/B-NAZANIN.TTF" ,width = 800, height = 800,
                      background_color ='white',
                      mask=mask,contour_color='#023075',
                      contour_width=3,colormap='rainbow',
                      stopwords=stopwords_list(),
                      min_font_size = 10).generate(ahmadinejad)
 
plt.figure()
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
# plt.tight_layout(pad = 0)
plt.savefig("mahmadi.jpeg", format="jpeg") 
plt.show()

```
