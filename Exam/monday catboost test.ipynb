استفاده کنید catboost راهنمایی: از کتابخانه

```python
from sklearn.model_selection import train_test_split
import category_encoders as ce
```

```python
X = data.drop('loss', axis = 1)
y = data['loss']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
# Define catboost encoder
cbe_encoder = ce.cat_boost.CatBoostEncoder()
  
# Fit encoder and transform the features
cbe_encoder.fit(X_train, y_train)
train_cbe = cbe_encoder.transform(X_train)
test_cbe = cbe_encoder.transform(X_test)
  
```

```python
train_cbe
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
model_linearRegression = make_pipeline(StandardScaler(), LinearRegression())
model_linearRegression.fit(train_cbe, y_train)
y_pred = model_linearRegression.predict(test_cbe)
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.svm import SVR
```

```python
svr=make_pipeline(StandardScaler(),SVR(kernel='rbf'))
svr.fit(train_cbe, y_train)
y_pred= svr.predict(test_cbe)
mean_squared_error(y_test, y_pred)
```
