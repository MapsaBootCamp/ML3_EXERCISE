```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open('BarakObama.txt', 'r',encoding="utf8") as file:
    Barak = file.read()
with open('GoergeBush.txt', 'r',encoding="utf8") as file:
    Bush = file.read()
with open('JohnFKennedy.txt', 'r',encoding="utf8") as file:
    Kennedy = file.read()
with open('MartinLutherKing.txt', 'r',encoding="utf8") as file:
    LutherKing = file.read()
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(Barak)
# plot the WordCloud image                      
plt.figure(figsize = (4, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
