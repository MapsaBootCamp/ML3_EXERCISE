```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
with open('BarakObama.txt','r') as f1:
  obama=f1.read()
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(obama)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
